AMOSTRA SISTEMÁTICA

```python
#importando ceil (arredonda para o maior)
from math import ceil
```

```python
#lendo csv
dataset = pd.read_csv('iris.csv')
dataset
```

```python
#criação de amostragem de o valor de k (valor que indica a contagem)
população = len(dataset)
amostra = 15
k = ceil(população/amostra)
k
```

```python
# valor randomico r para iniciar a amostra
r = np.random.randint(low= 1, high= k + 1, size= 1)
r
```

```python
#gerando escolhidos a partir do acumulador
acumulador = r[0]
escolhidos = []

for i in range(amostra):
  escolhidos.append(acumulador)
  acumulador += k

escolhidos
```

```python
test = dataset.loc[escolhidos]
test
```
