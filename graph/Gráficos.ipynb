```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
dataset = pd.read_csv("trees.csv")
dataset.head()
```

```python
#histograma pyplot
plt.hist(dataset.iloc[:,1], bins = 6, )
```

```python
#histograma com seaborn
sb.histplot(dataset.iloc[:,1])
```

```python
#histograma com seaborn 2
sb.distplot(dataset.iloc[:,1], bins=6, kde=True, color="blue", hist_kws={'edgecolor' : 'black'})
```

```python
#Dispersão com matplot
plt.scatter(dataset.Girth, dataset.Volume)
```

```python
#dispersão com seaborn
sb.regplot(dataset.Girth,dataset.Volume, data= dataset, fit_reg= False)
```

```python
#multiplos gráficos na mesma tela
plt.figure(1)
plt.subplot(2,2,1)
plt.scatter(dataset.Girth, dataset.Volume)
plt.subplot(2,2,2)
plt.scatter(dataset.Girth, dataset.Height)
plt.subplot(2,2,3)
plt.scatter(dataset.Volume, dataset.Height)
plt.subplot(2,2,4)
plt.hist(dataset.Volume)
```

```python
#boxplot com matplotlib
plt.boxplot(dataset.Volume, vert= False, patch_artist= True)
```
