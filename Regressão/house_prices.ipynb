```python
total = train.isnull().sum().sort_values(ascending= False)
percent = (train.isnull().sum()/train.isnull().count().sort_values(ascending= False))
nulos = pd.concat([total, percent], axis= 1, keys= ['Total', 'Percent'])
nulos.head(20)
```

```python
for data in fulldata:
    data.drop((nulos[nulos['Total'] > 1]).index,1, inplace= True)
    data.drop(data.loc[data['Electrical'].isnull()].index, inplace= True)
    
train.isnull().sum().max()
```

```python
from sklearn.preprocessing import LabelEncoder

for data in fulldata:
    for col in data.columns:
        data[col] = LabelEncoder().fit_transform(data[col])
```

```python
train.shape
```

```python
X = train.iloc[:,:62].values
y = train.iloc[:, 62].values
```

```python
from sklearn.model_selection import train_test_split

rep = 10

X_train_list, X_test_list, y_train_list, y_test_list = [], [], [], []

for i in range(rep):
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state= i)

    X_train_list.append(X_train)
    X_test_list.append(X_test)
    y_train_list.append(y_train)
    y_test_list.append(y_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

classifiers = [
    LinearRegression(),
    RandomForestRegressor(100),
    DecisionTreeRegressor()

]
```

```python
from sklearn.metrics import r2_score
import numpy as np

accs_mean = []
names = []
for clf in classifiers:
    names.append(clf.__class__.__name__)
    accs = []
    for i in range(rep):
        clf.fit(X_train_list[i],y_train_list[i])
        p = clf.predict(X_test_list[i])
        accs.append(r2_score(p, y_test_list[i]))
    accs_mean.append(np.mean(accs))

rank = pd.DataFrame(data= {'Names' : names , 'Accs' : accs_mean})
```
