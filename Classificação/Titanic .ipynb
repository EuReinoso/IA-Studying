```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
import statistics as sts
```

```python
#lendo data
titanic = pd.read_csv('train.csv')
titanic.head()
```

# FAZENDO O TRATAMENTO DOS DADOS

```python
#fazendo tratamento
titanic.isnull().sum()
```

```python
#Verificando NaNs
titanic.Age.describe()
```

```python
#Substituindo NaNs em Age
titanic['Age'].fillna(28, inplace = True)
titanic.Age.isnull().sum()
```

```python
#verificando moda em Embarked
titanic.Embarked.value_counts()
```

```python
#substituindo NaNs pela moda
titanic.Embarked.fillna('S', inplace= True)
titanic.Embarked.isnull().sum()
```

```python
#vsubstituindo NaNs de Cabine por -1
titanic.Cabin.fillna('A', inplace = True)
titanic.Cabin.isnull().sum()

titanic.Cabin.value_counts()
```

# Separando treino e teste


```python
previsores = titanic.iloc[:,2:]
classe = titanic.iloc[:,1]
```

```python
#convertendo categoricos para numericos
ctg = ['Name', 'Sex','Ticket', 'Cabin','Embarked'] 


previsores.loc[:,ctg[0]] = LabelEncoder().fit_transform(previsores.loc[:,ctg[0]])

previsores.loc[:,ctg[1]] = LabelEncoder().fit_transform(previsores.loc[:,ctg[1]])

previsores.loc[:,ctg[2]] = LabelEncoder().fit_transform(previsores.loc[:,ctg[2]])

previsores.loc[:,ctg[3]] = LabelEncoder().fit_transform(previsores.loc[:,ctg[3]])

previsores.loc[:,ctg[4]] = LabelEncoder().fit_transform(previsores.loc[:,ctg[4]])

previsores
```

```python
titanic.head(1)
```

```python
#diviindo train test
X_train, X_test, y_train, y_test = train_test_split(previsores,
                                                    classe,
                                                    test_size= 0.3,
                                                    stratify= classe,
                                                    random_state= 0)

X_train.head()
```

# Naive Bayes

```python
#treinando
nb = GaussianNB()
nb.fit(X_train, y_train)
```

```python
#testando 
previsoes = nb.predict(X_test)
```

```python
#verificando precisão
accuracy_score(previsoes, y_test)
```

```python
#teste com melhores atributos
X_train1 = X_train.iloc[:,[2,3,5]]
X_test1 = X_test.iloc[:,[2,3,5]]
```

```python
nb1 = GaussianNB()
nb1.fit(X_train1, y_train)
```

```python
previsoes1 = nb1.predict(X_test1)
```

```python
accuracy_score(previsoes1, y_test)
```

# Árvore de decisão

```python
tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)
```

```python
previsoes = tree.predict(X_test)
```

```python
accuracy_score(previsoes, y_test)
```

Fazendo novo treinamento só com os atributos importantes

```python
previsoes1 = tree1.predict(X_test1)
```
